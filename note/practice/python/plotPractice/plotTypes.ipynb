```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 10)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('x')
plt.show()

# Create a figure with two subplots
fig, axs = plt.subplots(2)

# Loop through the subplots
for i in range(2):
    # Generate two random variables
    x = np.random.rand(100)
    y = np.random.rand(100)
    
    # Plot the data in the subplot
    axs[i].scatter(x, y)
axs.set(xlim=(0, 0.8), xticks = np.arange(0.1, 0.8),
        ylim=(0, 0.8), yticks = np.arange(0.1, 0.8))
# Show the figure
plt.show()
```

Different Plot types:
1. pairwise data

```python
# scatter
plt.style.use('_mpl-gallery')
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
sizes = np.random.uniform(1, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors, vmax=100, vmin=0)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8), 
       ylim=(0, 8), yticks=np.arange(1, 8))
plt.show()

# bar
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

fig, ax = plt.subplots()
ax.bar(x, y, width = 1, edgecolor = "white", linewidth = 0.7)
plt.show()

# stem
fig, ax = plt.subplots()
ax.stem(x, y)
plt.show()

# fill between x, y1, y2
np.random.seed(1)
x = np.linspace(0, 8, 16)
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 0.5, len(x))

fig, ax = plt.subplots()
ax.fill_between(x, y1, y2, alpha = .5, linewidth = 0)
ax.plot(x, (y1 + y2)/2, linewidth = 2)
plt.show()

# stackplot
x = np.arange(0, 10, 2)
ay = [1, 1.25, 2, 2.75, 3]
by = [1, 1, 1, 1, 1]
cy = [2, 1, 2, 1, 2]
y = np.vstack([ay, by, cy])

fig, ax = plt.subplots()
ax.stackplot(x, y)
plt.show()

# stairs
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

fig, ax = plt.subplots()
ax.stairs(y, linewidth= 2.5)
plt.show()
```

2. statistical distributions

```python
# hist
plt.style.use('_mpl-gallery')

np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

fig, ax = plt.subplots()
ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")
ax.set(xlim=(0, 8), xticks=np.arange(1, 8), 
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))
plt.show()

# boxplot
np.random.seed(10)
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))

fig, ax = plt.subplots()
VP = ax.boxplot(D, positions=[2, 4, 6], widths=1.5, patch_artist=True, 
                showmeans=False, showfliers=False,
                medianprops={"color": "white", "linewidth": 0.5}, 
                boxprops={"facecolor": "C0", "edgecolor": "white", "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5}, 
                capprops={"color": "C0", "linewidth": 1.5})
plt.show()

# errorbar
np.random.seed(1)
x = [2, 4, 6]
y = [3.6, 5, 4.2]
yerr = [0.9, 0.2, 0.5]

fig, ax = plt.subplots()
ax.errorbar(x, y, yerr, fmt='o', linewidth=2, capsize=6)
plt.show()

# violinplot
np.random.seed(10)
D = np.random.normal((3, 5, 4), (0.75, 1.00, 0.75), (200, 3))

fig, ax = plt.subplots()
vp = ax.violinplot(D, [2, 4, 6], widths=2, 
                   showmeans=False, showmedians=False, showextrema=False)
for body in vp['bodies']:
    body.set_alpha(0.9)
plt.show()

# eventplot
x =  [2, 4, 6]
D = np.random.gamma(4, size=(3, 50))
fig, ax = plt.subplots()
ax.eventplot(D, orientation="vertical", lineoffsets=x, linewidths=0.75)
plt.show()

# hist2d
plt.style.use("_mpl-gallery-nogrid")
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000)

fig, ax = plt.subplots()
ax.hist2d(x, y, bins=(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1)))
ax.set(xlim=(-2, 2), ylim=(-3, 3))
plt.title("hist2d")
plt.show()

# hexbin
fig, ax = plt.subplots()
ax.hexbin(x, y, gridsize=20)
ax.set(xlim=(-2, 2), ylim=(-3, 3))
plt.title("hexbin")
plt.show()

# pie
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4), 
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
plt.title("pie")
plt.show()

# ecdf
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

fig, ax = plt.subplots()
ax.ecdf(x)
plt.title("ecdf")
plt.show()
```

```python
# imshow
plt.style.use('_mpl-gallery-nogrid')
X, Y = np.meshgrid(np.linspace(-3, 3, 24), np.linspace(-3, 3, 16))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

fig, ax = plt.subplots()
ax.imshow(Z)
plt.title("imshow")
plt.show()

# pcolormesh
x = [-3, -2, -1.6, -1.2, -.8, -.5, -.2, .1, .3, .5, .8, 1.1, 1.5, 1.9, 2.3, 3]
X, Y = np.meshgrid(x, np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

fig, ax = plt.subplots()
ax.pcolormesh(X, Y, Z, vmin=-0.5, vmax=1.0)
plt.title("pcolormesh")
plt.show()
# contour
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 7)

fig, ax = plt.subplots()
ax.contour(X, Y, Z, levels=levels)
plt.show()

# contourf
fig, ax = plt.subplots()
ax.contourf(X, Y, Z, levels=levels)
plt.show()

# barbs
X, Y = np.meshgrid([1, 2, 3, 4], [1, 2, 3, 4])
angle = np.pi / 180 * np.array([[15., 30, 35, 45],
                                [25., 40, 55, 60],
                                [35., 50, 65, 75],
                                [45., 60, 75, 90]])
amplitude = np.array([[5, 10, 25, 50],
                      [10, 15, 300, 60],
                      [15, 26, 50, 70],
                      [20, 45, 80, 100]])
U = amplitude * np.sin(angle)
V = amplitude * np.cos(angle)

fig, ax = plt.subplots()
ax.barbs(X, Y, U, V, barbcolor='C0', flagcolor='C0', length=7, linewidth=1.5)
ax.set(xlim=(0, 4.5), ylim=(0, 4.5))
plt.show()
```
